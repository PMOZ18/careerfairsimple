```python
!pip install pandas transformers openpyxl --quiet

import pandas as pd
from transformers import pipeline
from google.colab import files

# Upload the file
uploaded = files.upload()

# Get uploaded file name
file_name = list(uploaded.keys())[0]
print(f"File uploaded: {file_name}")

# Load zero-shot classifier
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# Load Excel file
df = pd.read_excel(file_name)

# Show available columns
print(df.columns)

# Adjust column names based on your file
company_col = "Company name"
roles_col = "Roles looking to offer"
domains_col = "Domains Interested"

# Target labels for classification
candidate_labels = ["technical role", "non-technical role"]

# List to keep companies that pass the filtering
final_companies = []

for index, row in df.iterrows():
    company = row[company_col]
    roles = str(row[roles_col])
    domains = str(row[domains_col])

    exclude = False

    # --- Step 1: Check roles (technical vs non-technical) ---
    role_list = [role.strip() for role in roles.split(',')]
    for role in role_list:
        if not role:
            continue
        result = classifier(role, candidate_labels)
        predicted_label = result['labels'][0]

        if predicted_label == "non-technical role":
            exclude = True
            break  # if even one non-technical role, exclude

    # --- Step 2: Check domains (MBA-only case) ---
    if not exclude:
        domains_upper = domains.upper()

        # Check if MBA is mentioned
        if 'MBA' in domains_upper:
            # Now check if 'Computer Science' is mentioned
            if 'COMPUTER SCIENCE' in domains_upper:
                # Count how many fields are listed
                domain_list = [d.strip() for d in domains.split(',')]
                if len(domain_list) <= 2:
                    # Only MBA + CS or MBA alone -> exclude
                    exclude = True
            else:
                # MBA mentioned but no Computer Science -> exclude
                exclude = True

    # --- Step 3: Add company if it passes all checks ---
    if not exclude:
        final_companies.append({
            'company': company,
            'roles': roles,
            'domains': domains
        })

# Console-style output
print("\n=== Final Companies after Strict Filtering ===\n")
print(f"Total companies filtered: {len(final_companies)}\n")

for company in final_companies:
    print(f"Company: {company['company']}")
    print(f"  Roles: {company['roles']}")
    print(f"  Domains: {company['domains']}\n")
```
